```python
import json
  
def load_from_file():
    # Opening JSON file
    with open('projects.json', 'r') as openfile:

        # Reading from json file
        json_object = json.load(openfile)

    return json_object
```

```python
projects = load_from_file()
```

```python
len(projects)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(projects.values())
```

```python
df['LotteryStageSummery'] = df['LotteryStageSummery'].astype(int)
df['LotteryApparmentsNum'] = df['LotteryApparmentsNum'].astype(int)
```

```python
df.head()
```

```python
import numpy as np

df_agg = df.groupby("CityDescription").agg({
    "LotteryApparmentsNum": [np.sum],
    "LotteryStageSummery": [np.max] } 
)
#df_agg["pct"] = df_agg["sum"]/df_agg["amax"]
df_agg
```

```python
df_agg.columns
```

```python
df_agg["pct"] = df_agg[('LotteryApparmentsNum', 'sum')]/df_agg[('LotteryStageSummery', 'amax')]*100
```

```python
df_agg.sort_values('pct', ascending=False)
```
